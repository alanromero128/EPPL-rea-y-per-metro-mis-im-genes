## Formula para calcular el area
$A=(B*H)/2$

---
## Datos de entrada
* $H$ ?
* $B$ ?

```python
# Calcular area de un triangulo
# Variables A,B,H
# Formulas A = (B*H)/2
# Capturar el valor de A
B = int(input("Ingrese el valor de la base: "))
H = int(input("Ingrese el valor de la altura: "))
A = (B*H)/2
print("El area del triangulo es: ", A)
```
